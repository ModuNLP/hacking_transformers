## Motivation

### Model Size
- Out of Memory: GPU 메모리 부족으로 모델 실행이 불가능함
- Training Time: BERT base의 경우 16개의 TPU(128GB)로 4일 걸림
- Model Degradation: BERT large보다 BERT xlarge가 더 안 좋음.


## BERT to ALBERT

### Factorized embedding parameterization
- General WordPiece Embedding -> BERT
    - V * E = 23,040K (V=30K, E=768)

- Factorized Embedding -> ALBERT
    - V * E + E * H = 3,077K (V=30K, E=100, H=768)
    - 기존 방식의 약 1/7 메모리 사용


- self.factorize = E*H
- Token Embedding -> Hidden

- self.linear_lm = H*E
- Hidden -> Token Embedding


### Cross-layer parameter sharing 
- General Layers -> BERT

- Cross-layer parameter sharing -> ALBERT


- General layers -> Cross-layer parameter sharing


### Sentence Order Prediction (SOP) 
- Next Sentence Prediction (NSP) -> BERT

- Sentence Order Prediction (SOP) -> ALBERT


- SOP로 학습한 모델은 NSP 성능도 좋음.
